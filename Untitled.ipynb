```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
from scipy import stats
import folium
import geopandas as gpd
from geopy.geocoders import Nominatim
from sklearn.cluster import DBSCAN, KMeans, MiniBatchKMeans, SpectralClustering
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
import xgboost
import shap
shap.initjs()

from sklearn.preprocessing import LabelEncoder
import collections.abc
from collections.abc import MutableSequence 
```

```python
X_injuries = pd.read_csv("PCA_injuries.csv", index_col=0)
```

```python
data = pd.read_csv("Cleaned data/Cleaned_Data.csv", index_col=0, parse_dates=['CRASH DATE'])
data['YEAR'] = data['CRASH DATE'].dt.year
data['MONTH'] = data['CRASH DATE'].dt.month
data['DAY'] = data['CRASH DATE'].dt.day
data['HOUR'] = data['CRASH TIME'].apply(lambda i:int(i.split(':')[0]))
```

```python
scores = []
print("Begin")
for k in range(1, 10):
    print(k, " clusters", end = "; ")
    scores.append([])
    for seed in range(3):
        print(seed)
        kmeans = MiniBatchKMeans(
            n_clusters=k,
            max_iter=25,
            n_init=1,
            random_state=seed,
            batch_size=10240
            )
        kmeans.fit(X_injuries)
        y_kmeans = kmeans.predict(X_injuries)

        scores[k-1].append(kmeans.inertia_)
```

```python
scores
```

```python
average_inertias = [sum(cluster_scores) / len(cluster_scores) for cluster_scores in scores]

# Prepare the x values (number of clusters)
cluster_range = range(1, 10)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(cluster_range, average_inertias, marker='o')

# Add titles and labels
plt.title('Elbow Method for KMeans Clustering')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Average Inertia')

# Show the plot
plt.grid(True)
plt.show()
```

```python
kmeans = MiniBatchKMeans(
    n_clusters=4,
    max_iter=25,
    n_init=1,
    random_state=42,
    batch_size=10240
    )

kmeans.fit(X_injuries)
y = kmeans.predict(X_injuries)
```

```python
y
```

```python
print(X_injuries.shape)
data = data[data['NUMBER OF PERSONS INJURED'] > 0]
print(data.shape)
```

```python
data['GROUP'] = y
```

```python
data.columns
data = data.drop(columns=['CRASH DATE', 'CRASH TIME', 'ON STREET NAME', 'OFF STREET NAME', 'CROSS STREET NAME', 'LOCATION'])
```

```python
non_encoded = data.copy()
```

```python
encoders = {}
for col in data.columns:
    if data[col].dtype == 'O':
        col_encoder = LabelEncoder()
        scaler = StandardScaler()
        data[col] = col_encoder.fit_transform(data[col])
    encoders[col] = col_encoder
```

```python
X_train = data[data['YEAR']<2023].drop(columns=['GROUP'])
X_test = data[data['YEAR']>=2023].drop(columns=['GROUP'])
y_train = data[data['YEAR']<2023]['GROUP']
y_test = data[data['YEAR']>=2023]['GROUP']
```

```python
d_train = xgboost.DMatrix(X_train, label=y_train)
d_test = xgboost.DMatrix(X_test, label=y_test)
```

```python
params = {
    "eta": 0.01,
    "objective": "multi:softprob",
    "subsample": 0.5,
    "base_score": np.mean(y_train),
    "eval_metric": "auc",
    "num_class": 4
}
model = xgboost.train(
    params,
    d_train,
    5000,
    evals=[(d_test, "test")],
    verbose_eval=100,
    early_stopping_rounds=20,
)
```

```python
xgboost.plot_importance(model)
plt.title("xgboost.plot_importance(model)")
plt.show()
```

```python
xgboost.plot_importance(model, importance_type="cover")
plt.title('xgboost.plot_importance(model, importance_type="cover")')
plt.show()
```

```python
xgboost.plot_importance(model, importance_type="gain")
plt.title('xgboost.plot_importance(model, importance_type="gain")')
plt.show()
```

```python
explainer = shap.TreeExplainer(model)
```

```python
data_sample = data.sample(n=30000)
d_sample = xgboost.DMatrix(data_sample.drop(columns=['GROUP']), label=data_sample['GROUP'])
```

```python
shap_values=explainer.shap_values(d_sample)
```
